### Data Importing

```sas
/** Import Data **/
FILENAME REFFILE '/folders/myfolders/sas_university_edition_vm/studenttest.txt';

PROC IMPORT DATAFILE=REFFILE
	DBMS=DLM
	OUT=WORK.Student;
	GETNAMES=NO;
RUN;

PROC CONTENTS DATA=WORK.Student; RUN;


/** Print Data **/
proc print DATA=WORK.Student; 
run;
```
